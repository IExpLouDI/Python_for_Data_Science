- С помощью функции plot постройте график, соединяющий линиями точки с горизонтальными координатами из списка x и вертикальными из списка y.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

%matplotlib inline
%config InlineBackend.figure_format = 'svg'
```

```python
x = np.arange(1, 8)
y = np.array([3.5, 3.8, 4.2, 4.5, 5, 5.5, 7])
x, y
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2)
ax1, ax2 = ax.flatten()

ax1.plot(x, y)
ax1.grid(color='lightblue')

ax2.scatter(x, y)
ax2.grid(color='lightblue')

fig.set_size_inches([8, 3.5])
plt.subplots_adjust(wspace=0.3)
```

- В левом верхнем графике для координат по вертикали используйте y1,в правом верхнем - y2, в левом нижнем - y3, в правом нижнем - y4.

```python
x = np.linspace(-3, 3, 51)
x
```

```python
y1 = x**2
y2 = 2 * x + 0.5
y3 = -3 * x + 0.5
y4 = np.sin(x)
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2)
a1, a2, a3, a4 = ax.flatten()

a1.plot(x, y1)
a1.set_title('График y1')
a1.set_xlim([-5, 5])
a1.grid()

a2.plot(x, y2)
a2.set_title('График y2')
a2.grid()

a3.plot(x, y3)
a3.set_title('График y3')
a3.grid()

a4.plot(x, y4)
a4.set_title('График y4')
a4.grid()

fig.set_size_inches([8, 6])
plt.subplots_adjust(wspace=0.3, hspace=0.3)

plt.show()
```
